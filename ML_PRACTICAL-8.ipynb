**-> IMPLEMENTATION OF SUPPORT VECTOR MACHINE(SVM) FOR LINEARLY SEPARABLE DATA.**

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the dataset
iris=pd.read_csv('IRIS.csv')

# Checking the dataset
iris.head()
```

```python
# Creating a pairplot to visualize the similarities and especially difference between the species
sns.pairplot(data=iris, hue='species', palette='Set1')
```

**TRAINING & TESTING DATA SPLIT :**

```python
from sklearn.model_selection import train_test_split

# Separating the independent variables from dependent variables
x=iris.iloc[:,:-1]
y=iris.iloc[:,4]

x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.30)
```

**TRAINING & FITTING THE MODEL :**

```python
from sklearn.svm import SVC
model=SVC(kernel='linear') 
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

**PREDICTIONS USING TRAINED MODEL:**



```python
y_pred=model.predict(x_test)
```

**CLASSIFICATION REPORT & CONFUSION MATRIX :**

```python
# Importing the classification report and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
