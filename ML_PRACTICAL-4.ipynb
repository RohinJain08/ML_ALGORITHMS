**Simple and Multiple linear regression using Normal equation with regularization.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Real_Estate_Price.csv')
df
```

```python
x=df.drop(['No','Y house price of unit area'],axis=1)

y=df['Y house price of unit area']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=2)

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_x,train_y)
```

```python
reg.score(test_x,test_y)
```

```python
reg.score(train_x,train_y)
```

***L1- LASSO REGULARIZATION***

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=0.001,max_iter=100000,tol=0.1)
lasso_reg.fit(train_x,train_y)
```

```python
lasso_reg.score(test_x,test_y)
```

```python
lasso_reg.score(train_x,train_y)
```


***L2- RIDGE REGULARIZATION***


```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=2)

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_x,train_y)
```

```python
reg.score(test_x,test_y)
```

```python
reg.score(train_x,train_y)
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=1,max_iter=100000,tol=0.1)
reg.fit(train_x,train_y)
```

```python
reg.score(test_x,test_y)
```

```python
reg.score(train_x,train_y)
```

```python
import pandas as pd
import numpy as np
import io
import math
```

```python
#reading the csv file and printing its data
data = pd.read_csv('Boston.csv')
print(data)
```

```python
#dividing the data into two parts 
X_train_d = data.iloc[:400,1:14]
X_test_d = data.iloc[400:,1:14]
Y_train_d = data.iloc[:400,-1]
Y_test_d = data.iloc[400:,-1]

#converting the split data into matrices and vectors
X_train = X_train_d.to_numpy()
X_test = X_test_d.to_numpy()
Y_train = np.reshape(Y_train_d.to_numpy(), (len(Y_train_d.to_numpy()), 1))
Y_test = np.reshape(Y_test_d.to_numpy(), (len(Y_test_d.to_numpy()), 1))

#Adding 1s column
X_train_col = np.ones((400,1))
X_test_col = np.ones((106,1))

#Combining the columns of 1s to the  matrices
X_train = np.concatenate([X_train_col,X_train], axis=1)
X_test = np.concatenate([X_test_col,X_test], axis=1)

#Normal equation Formula
X_train_transpose = [[X_train[j][i] for j in range(len(X_train))] for i in range(len(X_train[0]))]
theta = np.dot(X_train_transpose, X_train)
theta = np.linalg.inv(theta)
theta = np.dot(theta, X_train_transpose)
theta = np.dot(theta, Y_train)

#printing the derived formula 
print("formula for predicting the value of 'medv' = ", end = " ")
for i in range(13):
  print("(",theta[i], "*", data.columns[i+1],")", end = " ")
  if(i!=12):
    print("+", end=" ")
```

```python
#Error Finding
sum=0
for i in range(106):
  y_pred=0
  for j in range(13):
    y_pred = y_pred + (theta[j]*X_test[i][j])
  sum = sum + pow((y_pred-Y_test[i]),2)
sum=sum/106
error = math.sqrt(sum)
print("The error after testing the program is ", error)
```
