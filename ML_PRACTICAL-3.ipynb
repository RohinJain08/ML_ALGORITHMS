**1) Simple linear regression using Gradient Descent without regularization.(Without using sklearn or equivalent library).**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('Salary_Data.csv')

x= dataset['YearsExperience']
y=dataset['Salary']
plt.scatter(x,y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
m = 0
c = 0

alpha = 0.0001
n=float(len(x))

for i in range(1000):
   
    y_pred = m*x+c
    d_m = float((-2/n) * sum(x*(y-y_pred)))
    d_c = float((-2/n) * sum(y-y_pred))
    m = m-alpha * d_m
    c = c-alpha * d_c

y_pred = m*x + c
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)], color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()  
```

**2) Multiple linear regression using Gradient Descent without regularization.(Without using sklearn or equivalent library).**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('Real_Estate_Price.csv')
df
```

```python
df['No']=1
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data=scaler.fit_transform(df)
data=pd.DataFrame(data)
data
```

```python
data.iloc[:,0]=1
data
```

```python
x = data.iloc[0:,0:7]
y=data.iloc[0:,7]
y
```

```python

x=np.array(x)
y=np.array(y)
cp=[]
itr=[]

def cost(data,params):
    total_cost=0
    for i in range(414):
        total_cost += (1/414)*((data[i]*params).sum() - y[i])**2
    cp.append(total_cost)
    return total_cost 

def gd(data,params,lrate,iter_value):
    for i in range(iter_value):
        itr.append(i)
        slopes = np.zeros(7)
        for j in range(414):
            for k in range(7):
                slopes[k] += (1/414)*((data[j]*params).sum() - y[j])*data[j][k]
        params = params - lrate*slopes
        cost(data,params)
    return params  

params=np.zeros(7)
lrate=0.01
iter_value=200
params=gd(x,params,lrate,iter_value)
print(params)
print("\n")
plt.plot(itr,cp)
```
